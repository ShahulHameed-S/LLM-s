```python
docs=[
    "Gaming laptops are portable powerhouses designed to run visually stunning video games and resource-heavy software on the go",
    "Unlike standard laptops, they feature powerful dedicated graphics processors (GPUs) that render fast, high-quality visuals",
    "These devices are equipped with high-end CPUs, generous RAM, and fast solid-state drives (SSDs) for quick loading and seamless multitasking",
    "To maintain optimal performance during intense gameplay, they include advanced thermal management and cooling systems",
    "Many gaming laptops boast high-refresh-rate displays, ranging from 144Hz to 240Hz or higher, ensuring fluid and responsive animation",
    "Aesthetically, they often feature bold designs, aggressive cooling vents, and customizable RGB keyboards.",
    "Because of their heavy-duty components, they are typically thicker and heavier than standard ultrabooks",
    "Battery life is a common drawback, as intensive gaming drains the battery very quickly.",
    "While internal upgrades are usually limited to RAM and storage, their built-in hardware provides excellent longevity",
    "Despite these trade-offs, they offer unmatched versatility, making them ideal for both competitive gaming and professional tasks like 3D modeling"
]
```

```python
model=SentenceTransformer('all-MiniLM-L6-v2')
doc_embed=model.encode(docs).astype("float32")
```

```python
doc_embed.shape
```

```python
dimension=doc_embed.shape[1]
index=faiss.IndexFlatL2(dimension)
index.add(doc_embed)
```

```python
query="why we want to choose gaming laptops instead of normal one"
query_embed=model.encode([query]).astype("float32")
top_k=2
distance,indices=index.search(query_embed,top_k)#indices means number of sentence(documents in the docs),distance for similarity
retrieved_chunk=[docs[i] for i in indices[0]]
print("Retrieved Chunks")
for j in retrieved_chunk:
  print('-',j)
```

```python
context = "\n".join(retrieved_chunk)
prompt = f"""Answer the question based only on the context below.
Context: {context}
Question: {query}
Answer:"""

print("Final Answer:\n")

# Using the lightest and fastest model: gemini-2.5-flash
model_instance = genai.GenerativeModel(model_name='gemini-2.5-flash')
response = model_instance.generate_content(
    contents=prompt,
    generation_config=genai.types.GenerationConfig(
        temperature=0.4
    )
)

print(response.text)
```
