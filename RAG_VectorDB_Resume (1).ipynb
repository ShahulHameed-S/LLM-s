```python
model=SentenceTransformer('all-MiniLM-L6-v2')
doc_embed=model.encode(docs).astype("float32")
```

```python
dimension=doc_embed.shape[1]
index=faiss.IndexFlatL2(dimension)
index.add(doc_embed)
```

```python
query = "who is padma kumar"

embedding_model = SentenceTransformer('all-MiniLM-L6-v2') # Instantiate SentenceTransformer for embeddings
query_embed = embedding_model.encode([query]).astype("float32")

top_k = 2
distance, indices = index.search(query_embed, top_k)

retrieved_chunk = [docs[i] for i in indices[0]]

print("Retrieved Chunks\n")

for j in retrieved_chunk:
    lines = j.split("\n")

    for line in lines:
        line = line.strip()

        if line and not line.startswith("•"):
            print("-", line)

    print("\n----------------------\n")
```

```python

query="who is padma kumar"
query_embed=model.encode([query]).astype("float32")
top_k=2
distance,indices=index.search(query_embed,top_k)#indices means number of sentence(documents in the docs),distance for similarity
retrieved_chunk=[docs[i] for i in indices[0]]
print("Retrieved Chunks")
for j in retrieved_chunk:
  print('-',j)
```

```python
context = "\n".join(retrieved_chunk)
prompt = f"""Answer the question based only on the context below.
Context: {context}
Question: {query}
Answer:"""

print("Final Answer:\n")

# Using the lightest and fastest model: gemini-2.5-flash
model = genai.GenerativeModel('gemini-2.5-flash')
response = model.generate_content(
    contents=prompt,
    generation_config=genai.types.GenerationConfig(
        temperature=0.4
    )
)

print(response.text)
```
